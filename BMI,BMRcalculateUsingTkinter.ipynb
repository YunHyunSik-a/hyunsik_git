```python
from tkinter import *
window = Tk()

#사이즈와 제목넣기
window.geometry("400x250")
window.title("BMI 계산기")

#입력 위젯 작성
w_entry = Entry(window, width = 10)
w_entry.grid(row = 4, column = 3, padx = 10, pady = 10, columnspan = 2)
h_entry = Entry(window, width = 10)
h_entry.grid(row = 4, column = 5, padx = 10, pady = 10, columnspan = 2)
a_entry = Entry(window, width = 10)
a_entry.grid(row = 4, column = 7, padx = 10, columnspan = 2)

#BMI식, 기초대사량 식 함수정의
def BMIcalculate():
    weight = float(w_entry.get())
    height = float(h_entry.get())
    BMI = round(weight/height**2,2)
    bg_color = ""
    result = ""
    #비만도 판별
    if(BMI < 18.5):
        bg_color = "Green"
        result = "저체중"
    if(BMI >= 18.5 and BMI <= 22.9):
        bg_color = "Green"
        result = "정상"
    if(BMI >= 23.0 and BMI <= 24.9):
        bg_color = "orange"
        result = "과체중"
    if(BMI >= 25.0 and BMI <= 29.9):
        bg_color = "orange"
        result = "경도비만"
    if(BMI >= 30.0 and BMI <= 34.9):
        bg_color = "Red"
        result = "중도비만"
    if(BMI >=  35.0):
        bg_color = "Red"
        result = "고도비만"  
    BMI_label = Label(window, text= "BMI : {0} {1}".format(BMI,result) , bg = bg_color).grid(row = 8,column = 8)

#성별에 따른 기초대사량 계산식, 키는 m단위로 입력되었으므로 100을곱해줌

def gender_male():
    weight = float(w_entry.get())
    height = float(h_entry.get())
    age = int(a_entry.get())
    BMR = round(66.47 + (13.75 * weight) + (5 * height * 100) - (6.76 * age))
    bg_color = ""
    BMR_lable = Label(window, text = "기초대사량 : {0}".format(BMR),bg = "white").grid(row = 12, column = 8)
    
def gender_female():
    weight = float(w_entry.get())
    height = float(h_entry.get())
    age = int(a_entry.get())
    BMR = round(665.1 + (9.56 * weight) + (1.85 * height * 100) - (4.68 * age))
    bg_color = ""
    BMR_lable = Label(window, text= "기초대사량 : {0}".format(BMR),bg = "white").grid(row = 12, column = 8)
    
#위젯정리
w_label = Label(window, text = "몸무게(kg)").grid(row = 3, column = 3) 
h_label = Label(window, text = "키(m)").grid(row = 3, column = 5)
a_label = Label(window, text = "나이").grid(row = 3,column = 7)
BMI_label = Label(window, text = "BMI").grid(row = 6,column = 3)
calc_btn = Button(window, text = "계산하기",command = BMIcalculate).grid(row = 8,column = 3,padx = 10, pady = 10)

BMR_label = Label(window, text = "기초대사량 계산").grid(row = 10, column = 3)
#gender 정의 하기
gender_male = Button(window, text = '남자', command = gender_male).grid(row = 12, column=3, padx =10, pady=10)
gender_female = Button(window, text = '여자', command = gender_female).grid(row = 12, column=5, padx =10, pady=10)


window.mainloop()  
```
